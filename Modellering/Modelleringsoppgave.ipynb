# Oppgave 1

a er smitteraten. Modellen tar antall smittede og adderer antall folk de smittede har smittet

```python
from pylab import *

I = 3           #antall smittet
a = 0.2         #smitterate
tid = 48        #uker
pop =157759     #total populasjon

tid_liste = [0]
smittet_liste = [I]

for t in range(tid):
    I = I +a*I*(1-(I/pop))
    tid_liste.append(t)
    smittet_liste.append(I)

plot(tid_liste, smittet_liste)
show()
```

Smitte har eksponensiell vekst dersom vi ikke legger til begrensninger.
Hvis man øker a, vil populasjonen bli smittet fortere. Hvis man minker a, vil smitten øke i midre grad.  

# Oppgave 2

```python
from pylab import *

I = 3          #antall smittet
a = 0.4        #smitterate
S= pop - I     #mottakelige
tid =48        #uker
pop =157759    #total populasjon
tid_liste = [0]
smittet_liste =[I]
utsatte_liste = [S]


for t in range(tid):
    I = I +a*I*(1-(I/pop))
    S = pop-I
    utsatte_liste.append(S)
    tid_liste.append(t)
    smittet_liste.append(I)

plot(tid_liste, utsatte_liste, color= "hotpink", label = "mottakelige")
plot(tid_liste,smittet_liste, color = "blue", label = "smittet")
xlabel("uker")
ylabel("populasjon")
legend()
show()
```

Det som skjer er at når antall smittede øker, vil antall mottakelige minke.
Hvis a og I er store, blir alle smittet fortere

# Oppgave 3

```python
from pylab import *

I = 3        #antall smittet
a = 0.4     #smitterate
b = 0.04   #bedringsrate
S= pop - I
R = 0        #friske
tid =48      #uker
pop =157759  #total populasjon
tid_liste = [0]
smittet_liste =[I]
utsatte_liste = [S]
friske_liste=[R]


for t in range(tid):
    I = I +a*I*(1-(I/pop))-(b*I)
    R = R+b*I
    S = pop - (I+R)
    utsatte_liste.append(S)
    tid_liste.append(t)
    smittet_liste.append(I)
    friske_liste.append(R)

plot(tid_liste, utsatte_liste, color= "hotpink", label = "utsatte")
plot(tid_liste,smittet_liste, color = "lightskyblue", label = "smittet")
plot(tid_liste, friske_liste, color="plum", label="friske")
xlabel("uker")
ylabel("populasjon")
legend()
show()
```

Dette er det vi kom fram på egen hånd, men jeg tror den er litt feil siden grafen blir negativ. 

```python
from pylab import *

pop = 157759 # Individer
a=4/pop      # Smitterate
b=3.5       # Bedringsrate
t=48         # Tid

I =3         # Smittede
S= pop-I     # Mottakelige
R =0         # Restituert/friske

syke = [I]
mulige =[S]
friske = [R]
t = [0]

for i in range(1, tid):
    S = S-a*S*I
    I = I+a*S*I- b*I
    R = R +b*I
    syke.append(I)
    mulige.append(S)
    friske.append(R)
    t.append(i)
    
plot(t, syke, label="smittet")
plot(t, mulige, label="mottakelige")
plot(t, friske, label = "friske")
legend()
show()
```

Smitteraten er veldig lav og det blir nesten ikke synlig. Mottakelige og friske er avhengig av hverandre og øker/minker speilvendt med hverandre.

```python
from pylab import *

pop = 157759 # Individer
a=4/pop      # Smitterate
b=3.5        # Bedringsrate
t=48         # Tid
v = 2      #vaksinerate

I =3         # Smittede
S= pop-I     # Mottakelige
R =0         # Restituert/friske

syke = [I]
mulige =[S]
friske = [R]
t = [0]

for i in range(1, tid):
    S = S-a*S*I
    I = (I+a*S*I- b*I)*v
    R = R +b*I
    syke.append(I)
    mulige.append(S)
    friske.append(R)
    t.append(i)
    
plot(t, syke, label="smittet")
plot(t, mulige, label="mottakelige")
plot(t, friske, label = "friske")
legend()
show()
```
