### 7.3

```python
from pylab import *

år = 20
planter = []
endring = []
tid = []

for t in range(år):
    N = 0.2*exp(t)+90
    planter.append(N)
    tid.append(t)
    
for i in range(len(tid)-1):
    dx = tid[i+1] - tid[i]
    dy = planter[i+1]- planter[i]
    der = dy/dx
    endring.append(der)

plot(tid, planter, color = "hotpink", label = "antall planter")
plot(tid[:-1],endring, color = "blue", label = "endringen i antall planter")
xlabel("år")
ylabel("planter")
legend()
show()
```

### 7.9

```python
from pylab import *

def der(f, x, dx):
    fder = (f(x + dx) - f(x)) / dx
    return fder

def temp(t):
    T = 70 * exp(-0.065*t)
    return T

tid = linspace(0, 60, 1000)       
dt = 1e-8
derivert = der(temp, tid, dt)
temperatur = temp(tid)

plot(tid, derivert, color= "hotpink", label = "Temperatur")
plot(tid, temperatur, color = "lightblue", label = "Endring i temperatur")
xlabel("minutter")
ylabel("temperatur (celcius)")
legend()
show()
print(f"temperaturen synker med {derivert[41]} celcius ved element nr. 42")
    
```
